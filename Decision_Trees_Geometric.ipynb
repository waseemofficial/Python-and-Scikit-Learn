```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from  sklearn import tree
from dtreeviz.trees import *
import dtreeviz
from IPython.display import Image,display_svg,SVG
%matplotlib inline
```

```python
# def analyzer(max_depth):
#     data=pd.read_csv
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/bhattbhavesh91/multivariate-adaptive-regression-splines-demo/main/Data.csv")
df.head(5)
```

```python
df.shape
```

```python
X = df[["TV", "radio", "newspaper"]]
y = df["sales"]
```

```python
regressor = DecisionTreeRegressor(random_state = 2022, max_depth=3) 
```

```python
regressor.fit(X, y)
```

```python
fig = plt.figure(figsize=(10,20))
_ = tree.plot_tree(regressor,
                   feature_names = ["TV", "radio", "newspaper"],
                   class_names = ["sales"],
                   filled=True)
     
```

```python
viz = dtreeviz.model(regressor,
               X,
               y,

               target_name='sales',
               feature_names=["TV", "radio", "newspaper"],
               class_names=[]
               )
```
